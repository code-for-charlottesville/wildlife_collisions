```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("WCV Collision Data.csv")
df['Date'] = pd.to_datetime(df['Date Admitted'], format='%m/%d/%Y')
df["Month"] = df["Date"].dt.month_name()
df["Year"] = df["Date"].dt.year
df["% Full Moon"] = 1739*["TBD"]

df = df.drop(labels=["Organization Name","Case Number", "Patient ID","Date Admitted","Rescue State","Circumstances of Rescue"], axis=1)
#df.head()
```

```python
months=["January","February","March","April","May","June","July","August","September","October","November","December"]
monthLabels = ["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
years=[2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

fig, ax = plt.subplots(nrows=1, ncols=2, sharex=False, sharey=False, figsize=(15,8))
plt.subplots_adjust(wspace=.35)
sns.countplot(data=df, x="Month", color="#2b8cbe",order=months, ax=ax[0]).set_xticklabels(labels=monthLabels,rotation=45)
sns.countplot(data=df, x="Year", color='#2b8cbe',order=years, ax=ax[1]).set_xticklabels(labels=years,rotation=45)
fig.suptitle("Calendar Trends (all animal types)")
fig.show()
```

```python

# Birds
owls = [
    "Eastern Screech-Owl",
    "Barred Owl",
    "Great Horned Owl",
    "Barn Owl"
]

hawks = [
    "Red-shouldered Hawk",
    "Cooper's Hawk",
    "Sharp-shinned Hawk",
    "Red-tailed Hawk",
    "Broad-winged Hawk",
    "American Kestrel"
]

birds_of_prey = [
    "Black Vulture",
    "Turkey Vulture",
    "Bald Eagle",
    "Osprey",
    "Merlin"
] + owls + hawks

songbirds_perching_birds = [
    "Eastern Bluebird",
    "American Robin",
    "Northern Cardinal",
    "Cedar Waxwing",
    "Indigo Bunting",
    "Rose-breasted Grosbeak",
    "Gray Catbird",
    "Tufted Titmouse",
    "Northern Mockingbird",
    "Baltimore Oriole",
    "Brown Thrasher",
    "House Finch",
    "Purple Finch",
    "Common Grackle",
    "House Sparrow",
    "Purple Martin",
    "White-throated Sparrow",
    "Chipping Sparrow",
    "Fox Sparrow",
    "Dark-eyed Junco",
    "Yellow-billed Cuckoo",
    'Hermit Thrush',
    'Wood Thrush',
    'Gray Kingbird',
    'American Goldfinch',
    'American Crow',
    'Fish Crow',
    'Carolina Wren',
    'Eastern Kingbird',
    'Blue Jay',
    'Red-eyed Vireo',
    'Common Raven',
    'Eastern Phoebe'
]

swallows_swifts = [
    "Barn Swallow",
    "Chimney Swift",
    "Northern Rough-winged Swallow",
    "Tree Swallow"
]

waterfowl_shorebirds = [
    "Mallard",
    "Canada Goose",
    "Wood Duck",
    "Hooded Merganser",
    "Lesser Scaup",
    "Ring-necked Pheasant",
    "Ring-billed Gull",
    "Sora",
    "Domestic Duck",
    'Double-crested Cormorant',
    'Great Blue Heron',
    'Muscovy Duck',
    'Virginia Rail'
]

others_birds = [
    "European Starling",
    "Rock Dove (Pigeon)",
    "Rock Dove",
    "Mourning Dove",
    'Ring-necked Dove',
    "Yellow-bellied Sapsucker",
    "Downy Woodpecker",
    "Hairy Woodpecker",
    "Red-bellied Woodpecker",
    "Red-headed Woodpecker",
    "Pileated Woodpecker",
    'Wild Turkey',
    'Northern Bobwhite',
    'Northern Flicker',
    'Whip-poor-will'
]

# Mammals
rodents_squirrels = [
    "Eastern Gray Squirrel",
    "eastern gray squirrel",
    "Eastern Fox Squirrel",
    "eastern fox squirrel",
    "Eastern Chipmunk",
    'eastern chipmunk'
]

carnivores = [
    "Virginia Opossum",
    "Red Fox",
    "Raccoon",
    "American Mink",
    "Gray Fox",
    "Striped Skunk",
    "American Black Bear",
    'American mink'
]

rabbits_and_hares = [
    "Domestic Rabbit",
    'Eastern Cottontail'
]

deer = [
    "White-tailed Deer", 
    "white-tailed deer"]

groundhog = [
    "Woodchuck (Groundhog)",
              "woodchuck"]

bats = [
    "Eastern Red Bat", 
    "Big Brown Bat",
    'eastern red bat',
    'big brown bat']

mice = [
    "White-footed Mouse",
    'white-footed mouse']

# Reptiles
turtles = [
    "Yellow-bellied Slider",
    "Eastern Painted Turtle",
    "Eastern Mud Turtle",
    "Eastern River Cooter",
    "Painted Turtle",
    "Common Musk Turtle",
    "Common Snapping Turtle",
    'common snapping turtle',
    'common box turtle',
    "Red-eared Slider",
    "Bog Turtle",
    "Coastal Plain Cooter",
    "Northern Red-bellied Cooter",
    "Eastern Box Turtle",
    "Snapping Turtle",
    'eastern mud turtle'
]

snakes = [
    "Eastern Ratsnake",
    "Eastern Hog-nosed Snake",
    "Dekay's Brown Snake",
    "Northern Black Racer",
    "Common Garter Snake"
]

# Amphibians
frogs_toads = [
    "Eastern American Toad",
    "Fowler's Toad",
    "American Toad"
]



# Giving categories for Small Groups (tentatively called genus to give me a label, not sure the technically right terms we will want to use)
genus_dict={
        "Owls":owls, 
        "Hawks":hawks, 
        "Birds of Prey":birds_of_prey, 
        "Songbirds and Perching Birds":songbirds_perching_birds, 
        "Swallows and Swifts":swallows_swifts,
        "Waterfowl and Shorebirds":waterfowl_shorebirds,
        "Other Birds":others_birds,

        "Rodents and Squirrels": rodents_squirrels,
        "Carnivore Mammals":carnivores,
        "Rabbits and Hares":rabbits_and_hares,
        "Deer":deer,
        "Groundhogs": groundhog,
        "Bats": bats,
        "Mice": mice,

        "Turtles":turtles,
        "Snakes":snakes,

        "Frogs and Toads":frogs_toads
    }

inverted_genus_dict={}
for small_group, genus_list in genus_dict.items():
    for genus in genus_list:
        inverted_genus_dict[genus] = small_group

for index,row in df.iterrows():
    df.loc[index, "Small Group"] = inverted_genus_dict[row["Common Species Name"]] 


# Giving categories for Large Groups (tentatively called family to give me a label, not sure the technically right terms we will want to use)

birds = birds_of_prey + songbirds_perching_birds + swallows_swifts + waterfowl_shorebirds + others_birds
mammals = rodents_squirrels + carnivores + rabbits_and_hares + deer + groundhog + bats + mice
reptiles = turtles + snakes
amphibians = frogs_toads

family_dict = {
    "Birds":birds, 
    "Mammals":mammals, 
    "Reptiles":reptiles, 
    "Amphibians":amphibians}

inverted_family_dict = {family: large_group for large_group, family_list in family_dict.items() for family in family_list}

# inverted_family_dict = {}
# for large_group, family_list in class_dict.items():
#     for family in family_list:
#         inverted_dict[family] = large_group

#Giving large group categories
for index,row in df.iterrows():
    df.loc[index, "Large Group"] = inverted_family_dict[row["Common Species Name"]]  
```

```python
# Overview trends

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,6))
graph1 = sns.countplot(df, x="Month", hue="Large Group", order=months, ax=ax1)
graph1.set_xticklabels(labels=monthLabels,rotation=45)
graph1.set_title("Monthly Collisions by Species Category")

graph2 = sns.countplot(df, x="Year", hue="Large Group", order=years, ax=ax2)
graph2.set_xticklabels(labels=years,rotation=45)
graph2.set_title("Yearly Collisions by Species Category")
plt.tight_layout()
```

```python
# General trends by animal type. Note that the y-axes are not the same (especially amphibians, with only 5 instances)
def make_plot(group_name, axisL, axisR):
    m = sns.countplot(data=df[df["Large Group"]==group_name], x="Month", color="#2b8cbe",order=months, ax=axisL)
    m.set_xticklabels(labels=monthLabels,rotation=45)
    m.title.set_text(f"{group_name} by Month")
    m.set_xlabel('')
    
    y = sns.countplot(data=df[df["Large Group"]==group_name], x="Year", color='#2b8cbe',order=years, ax=axisR)
    y.set_xticklabels(labels=years,rotation=45)
    y.title.set_text(f"{group_name} by Year")
    y.set_xlabel('')

#group_names = list of 4 families
def make_grid(group_names):
    fig, ax = plt.subplots(nrows=4, ncols=2, sharex=False, sharey=False, figsize=(10,10))
    for i, group in enumerate(group_names):
        make_plot(group, axisL = ax[i][0], axisR=ax[i][1])
    plt.tight_layout()
    #fig.suptitle("Calendar Trends")
    fig.show()

large_group_names = ["Mammals","Birds","Reptiles","Amphibians"]

make_grid(large_group_names)
```
